```python
import matplotlib.pyplot as plt
import numpy as np

def yearly_progress(df, y_features, x_feature):

    plt.figure(figsize=(15, 8))

    
    colors = plt.cm.get_cmap('tab10', len(y_features))  
    linestyles = ['-', '--', '-.', ':']  
    
    
    linestyles = linestyles * (len(y_features) // len(linestyles) + 1)  
    
    
    for i, feature in enumerate(y_features):
        plt.plot(df[x_feature], df[feature], 
                 color=colors(i), 
                 linestyle=linestyles[i], 
                 label=feature)  

    plt.legend(loc='best')
    plt.xlabel(x_feature)
    plt.ylabel('Feature values')
    plt.title(f'Yearly Progress of Features vs {x_feature}')
    plt.show()
```

```python
num_columns = df.columns[1:]
yearly_progress(df, num_columns, "Year")
```
